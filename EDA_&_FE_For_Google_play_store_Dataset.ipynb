## INSIGHTS AND OBSERVATION

DATA CLEANING

```python
df['Reviews'].unique()
```

```python
df['Reviews'].str.isnumeric().sum()
```

```python
df[~df['Reviews'].str.isnumeric()]
```

```python
df_copy=df.copy()
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
## COnvert 'Reviews' to numerical
df_copy['Reviews']=df_copy['Reviews'].astype('int')
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
## Convert 'M to 'K

df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Size']
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
chars_to_remove = ['$', ',', '+']
cols_to_clean = ['Installs', 'Price']
```

```python
for item in chars_to_remove:
    for col in cols_to_clean:
        df_copy[col] = df_copy[col].str.replace(item, '')
```

```python
df_copy.head()
```

```python
df_copy['Installs'].unique()
```

```python

df_copy['Price'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype('int')
df_copy['Price']=df_copy['Price'].astype('float')
```

```python
df_copy.info()
```

```python
df['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
df_copy.head()
```

```python
df_copy.to_csv('cleaned_google_play_store_data.csv',index=False)
```

## Explore the data

```python
# The the numerical and the categorical features in the dataset
numerical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']

print('We have {} numerical features : {}'.format(len(numerical_features), numerical_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
# Proprtion of count data on the categorical data
for col in categorical_features:
    print(df_copy[col].value_counts(normalize=True)*100)
    print('#'*50)
```

```python
# Visualize the proprtion of count data on the numerical data
plt.figure(figsize=(18, 12))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(categorical_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df_copy[numerical_features[i]], shade=True, color='r')
    plt.xlabel(numerical_features[i])
    plt.tight_layout()
```

## OBSERVATIONs
- Rating and Year are left skewed while Reviews, Size, Installs and Price are right skewed

```python
# Visualize the proprtion of count data on the numerical data
plt.figure(figsize=(18, 12))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
category =['Type','Content Rating']

for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=df_copy[category[i]], palette='Set2')
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```
